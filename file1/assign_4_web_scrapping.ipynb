```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
name=[]
datatype=[]
task=[]
att_type=[]
year=[]
noi=[]
noa=[]
start = 0
end = 4
expand = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')

for i in range(start,end):
    expand = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
    expand.click()
    time.sleep(3)
    try:
        for i in driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/h2/a'):
            name.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        name.append('-')
        
    
    try:
        for i in driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][2]/span'):
            datatype.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        datatype.append('-')
        
    
    try:
        for i in driver.find_elements(By.XPATH,'//div[1][@class="col-span-3 flex items-center gap-2"]/span[1]'):
            task.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        task.append('-')
        
    
    try:
        for i in driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[2]'):
            att_type.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        att_type.append('-')
        
    
    try:
        for i in driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[3]'):
            year.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        year.append('-')
        
    
    try:
        for i in driver.find_elements(By.XPATH,'//div[3][@class="col-span-3 flex items-center gap-2"]/span'):
            noi.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        noi.append('-')
        
    
    try:
        for i in driver.find_elements(By.XPATH,'//div[4][@class="col-span-3 flex items-center gap-2"]/span'):
            noa.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        noa.append('-')
    
    
    
  
    
    try:
        next_button = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
        driver.execute_script("arguments[0].click();", next_button)
    except Exception as e:
        print("An error occurred: ", e)

    
    time.sleep(5)
    expand.click()
    
    
    
    
        
    
        
```

```python
fixtures = driver.find_element(By.XPATH,'/html/body/header/div[3]/div[1]/ul/div[1]/a[2]')
fixtures.click()
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
delay=10
series=[]
g_name=[]
date=[]
times=[]
place=[]
for i in range(0,6):
    try:
        for i in driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]'):
            series.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        for i in driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]'):
            series.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,'//div[@class="match-place ng-scope"]//span[1]'):

            g_name.append(i.get_attribute('textContent'))
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        for i in driver.find_elements(By.XPATH,'//div[@class="match-place ng-scope"]//span[1]'):

            g_name.append(i.get_attribute('textContent'))
            

    line = driver.find_elements(By.XPATH, '//div[@class="match-place ng-scope"]')

    for i in line:
        try:
            plc = i.find_element(By.XPATH, './/span[@class="ng-binding"]').text
        except NoSuchElementException:
            plc = '-'
        place.append(plc)

    try:
        for i in driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]'):
            date.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        for i in driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]'):
            date.append(i.text)
    try:
        for i in driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]'):
            times.append(i.text)
    except (StaleElementReferenceException, NoSuchElementException ,TimeoutException) as e:
        for i in driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]'):
            times.append(i.text)
        
        
    try:
        
        more_button = driver.find_element(By.XPATH, '/html/body/section/div/div/div/div/div/div[2]/div[2]/div[2]/div/button')
        driver.execute_script("arguments[0].click();", more_button)
        time.sleep(2)
    except NoSuchElementException:
        break

    
        
    
```
